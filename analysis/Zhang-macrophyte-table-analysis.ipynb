## V2

```python
## load data v2
# pth = '/Users/ethankyzivat/Google Drive/Research/Lab Meetings/UAVSAR_paper_May_2021/Zhang-global-loss-veg-lakes-table.xlsx'
pth = '/mnt/g/Other/Zhang2017LakeMacrophytes/ZhangMacrophyteDecline.xlsx'
df = pd.read_excel(pth, na_values='NaN', skiprows=0)
df.reset_index(col_level=0, inplace=True)
# df.set_index('index', inplace=True) # doesn't work
df.drop(0, axis=0, inplace=True)
df.head(8)
```

```python
## Fill NA with most recent obs

cols_to_backfill = ['Reference','Site No.','Study sites', 'Country', 'Types', 'Area (km2)', 'Change trend', 'Area difference between final and initial year (km2)', 'Observation method']
df.loc[:, cols_to_backfill] = df[cols_to_backfill].fillna(method='ffill')
df.rename(columns={'index':'reference'}, inplace=True)
df.index.rename('Idx')
df.head(8)
```

```python
## Fix lake area values 
out=[]
for i in df['Area (km2)']:
    out.append(str(i).replace('>50', '60').replace('<10','1').replace('30-50', '40').replace('20-50', '35'))
df['Area (km2)'] = np.array(out, dtype='float')
df.head()
```

```python
## Remove any entries with no macrophyte % value
df.dropna(subset=['Aquatic vegetation cover (%)'], inplace=True)
```

```python
# joinby
# df.groupby('Study sites').last().sort_values('reference')
dfg = df.groupby('Study sites')
print('Mean of earliest M%: {}'.format(dfg['Aquatic vegetation cover (%)'].first().mean()))
print('Mean of most recent M%: {}'.format(dfg['Aquatic vegetation cover (%)'].last().mean()))
print('Mean of temporal mean M%: ' + str(dfg['Aquatic vegetation cover (%)'].mean().mean()))

print('\nWeighted mean of first M%: ' + str(np.average(dfg['Aquatic vegetation cover (%)'].first(), weights = dfg['Area (km2)'].mean())))
print('Weighted mean of last M%: ' + str(np.average(dfg['Aquatic vegetation cover (%)'].last(), weights = dfg['Area (km2)'].mean())))
print('Weighted mean of temporal mean M%: ' + str(np.average(dfg['Aquatic vegetation cover (%)'].mean(), weights = dfg['Area (km2)'].mean())))
```

# Save output

```python
## Save
pth_out=os.path.join(os.path.dirname(pth), 'Zhang_tabular_out.xlsx')
df.to_excel(pth_out)

pth_out_group_mean=os.path.join(os.path.dirname(pth), 'Zhang_tabular_grouped_mean_out.xlsx')
dfg.mean().to_excel(pth_out_group_mean)

print('saved')
```
